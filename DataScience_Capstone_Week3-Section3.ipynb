## Previous Code to Generate the DataFrame with Latitude and Longitude

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')

#########################

from pandas.io.html import read_html

page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

wikitables = read_html(page, attrs={'class':"wikitable sortable"})
print("Extracted {num} wikitables".format(num=len(wikitables)))
#wikitables



df = pd.DataFrame()
df1 = pd.DataFrame()

df = wikitables[0]

df1 = df[df.Borough != 'Not assigned']
df1.reset_index(drop=True, inplace=True)
#df1

#df1.loc[df['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = df1.loc[df['Borough']]
#df1.Neighbourhood.replace(df['Neighbourhood'] == 'Not assigned',df.Borough,inplace=True)

df1['Neighbourhood'] = np.where(df1['Neighbourhood'] == 'Not assigned', df1['Borough'], df1['Neighbourhood'])
#df1


df1 = df1.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
#df1

Final_result = df1.sort_values(by=['Borough'])
Final_result.reset_index(drop=True, inplace=True)
#Final_result

lat_lang = pd.DataFrame()
lat_lang = pd.read_csv('E:\Coursera\Capstone\Geospatial_Coordinates.csv')
#lat_lang.head(5)


lat_lang.rename(columns = {"Postal Code": "Postcode"}, inplace = True) 

#df_merge_col = pd.merge(Final_result, lat_lang, left_on='Postcode', right_on='Postal Code', how='outer')
df_merge_col = pd.merge(Final_result, lat_lang, on='Postcode')
df_merge_col
```

### Create a map of Canada, Toronto with neighborhoods superimposed on top

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Canada, Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)


# add markers to map
for lat, lng, borough, neighborhood in zip(df_merge_col['Latitude'], df_merge_col['Longitude'], df_merge_col['Borough'], df_merge_col['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  

map_toronto
```
